```python
global count,name,directory_present
count = 0
name = None
directory_present = False

def save_frame(frame, u_name, directory_present) :
    """This function is used to check for existence of identified Person named directory . If Not
       present new directory is created with the persons name. Also the name of person is checked 
       if the name is None then the name is also taken as input from the user. """
    import os
    
    # check Name
    if u_name is None :
            str = input("Plz Enter The Face Name To Store It With : ")
            u_name = str
    print("Directory Present : ",directory_present)
    
    # get Root Path
    thisdir = os.getcwd()
    check_path = thisdir + '/image_assets/' + u_name
    print(check_path)
    
    # Boolean Check if Directory Already Exists  
    if directory_present :
        
            # if Present Write face to the Directory
            cv.imwrite( check_path + "/" + u_name + '%d.jpg' % count, frame)
    else :
        
        # Check for existence of Directory in File System
        if not os.path.exists(check_path) :
            print("Face Directory Absent. Recreating Folder with Name ",u_name)
            os.mkdir(check_path)
            directory_present = True
        else :
            print("Face Directory Already Present.")
            directory_present = True
    return (u_name,directory_present)
```
