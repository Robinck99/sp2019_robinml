```python

predictor = dlib.shape_predictor('shape_predictor.dat')

def shape_to_np(shape, dtype="int"):
    # initialize the list of (x, y)-coordinates
    coords = np.zeros((68, 2), dtype=dtype)
 
    # loop over the 68 facial landmarks and convert them
    # to a 2-tuple of (x, y)-coordinates
    for i in range(0, 68):
        coords[i] = (shape.part(i).x, shape.part(i).y)
 
    # return the list of (x, y)-coordinates
    return coords

def face_landmarks(face_gray, face_frame) :
    shape = predictor(face_gray, face_frame)
    shape = shape_to_np(shape)
    for (x, y) in shape:
        cv.circle(face_frame, (x, y), 1, (0, 0, 255), -1)


def rect_boundary(rect):
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y
    return (x, y, w, h)
```

```python
import cv2 as cv
import dlib
import numpy as np

cam = cv.VideoCapture(0)
count = 0
# face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
# eye_cascade = cv.CascadeClassifier('haarcascade_eye.xml')

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor.dat')
coords = np.zeros((68, 2), dtype=int)

while True:
    _, frame = cam.read()
    gray_frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    print(cam.get(cv.CAP_PROP_FPS))
    
#     faces = face_cascade.detectMultiScale(gray_frame, 1.3, 5)
    
    face_frame = frame
    face_validation = False
    
    rects = detector(gray_frame,1)
    
    for (i, rect) in enumerate(rects) :
        cv.rectangle(frame, (rect.left(), rect.top()), (rect.right(), rect.bottom()), (0, 255, 0), 2)
        face_frame = frame[rect.top() : rect.bottom(), rect.left() : rect.right()]
        face_validation = True
        shape = predictor(gray_frame, rect)
        for i in range(0, 68):
            coords[i] = (shape.part(i).x, shape.part(i).y)
        
        for (x, y) in coords:
            cv.circle(frame, (x, y), 1, (0, 0, 255), -1)
            
    cv.imshow("frame", frame)
    if face_validation :
        cv.imshow("faces",face_frame)
#         name,directory_present = save_frame(face_frame,name,directory_present)
        count += 1
    
    if count > 30 :
        count = 0
    
    if cv.waitKey(1) & 0xFF == ord("q") :
        break
cam.release()
cv.destroyAllWindows()
```
