```python
state = 'goa'
data = pd.DataFrame()
for x in range(20):
    d = record[x][record[x].SUBDIVISION == state]
    d = d.iloc[:,[1,14]]
    data = data.append(d)
data.head()
```

```python
years_list = list(data.YEAR)

plt.style.use('ggplot')
plt.figure(figsize=(20,10))
plt.axis([1900,1921, 0.0, 4000.0])
plt.plot(data.YEAR, data.iloc[:,1])
plt.title('JAN 1901 - 1951 ')
plt.ylabel('Value in mm')
plt.xlabel(' Years')
plt.xticks(years_list)
plt.legend()
plt.show()
```
