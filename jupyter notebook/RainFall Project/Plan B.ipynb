```python
import pandas as pd
import os
import indian_state_list as isl

d_state = isl.d_state

months = [ 'YEAR','JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANNUAL']

path = os.getcwd()
record = []
for i in range(100):
    database = pd.read_csv(path + "/year_record_filtered/record_" + str(1901 + i) + "_filtered.csv")
    database = database.iloc[:,1:]
    record.append(database)
```

```python
def get_all_year_statewise(state):
    parent_db = pd.DataFrame()
    for i in range(100):
        d = record[i][record[i].SUBDIVISION == state]
        parent_db = parent_db.append(d.iloc[:,1:15])
    parent_db.reset_index(drop=True, inplace=True)
    return parent_db

pb_db = get_all_year_statewise('punjab')
mp_db = get_all_year_statewise('madhya pradesh')
pb_db.head()
```

```python
from matplotlib import pyplot as plt 
plt.style.use('ggplot')
plt.figure(figsize=(20,10))
plt.plot(pb_db.iloc[10,1:13])
plt.plot(mp_db.iloc[10,1:13])
plt.legend(['PB','MP'])
plt.show()
```
