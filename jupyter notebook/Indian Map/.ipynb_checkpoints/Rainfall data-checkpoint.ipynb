```python
import pandas as pd
import numpy as np
import indian_map_image_clor as in_map

d_state = {'rajasthan': 0, 'madhya pradesh': 1, 'maharashtra': 2, 'jammu & kashmir': 3, 'uttar pradesh': 4, 'andhra pradesh': 5, 'gujarat': 6, 'karnataka': 7, 'orissa': 8, 'chhattisgarh': 9, 'tamil nadu': 10, 'bihar': 11, 'arunachal pradesh': 12, 'assam': 13, 'jharkhand': 14, 'west bengal': 15, 'himachal pradesh': 16, 'uttarakhand': 17, 'punjab': 18, 'haryana': 19, 'kerala': 20, 'manipur': 21, 'meghalaya': 22, 'mizoram': 23, 'nagaland': 24, 'tripura': 25, 'sikkim': 26, 'goa': 27, 'delhi': 28}

parent_database = pd.read_csv('Sub_Division_IMD_2017.csv')
if parent_database.shape[0] > 0 :
    print("Dataset Read Complete")
    print(parent_database.head(5))
else :
    print("Dataset Read Incomplete, Error in Csv File.")
```

```python
data = parent_database[parent_database.YEAR == 2017]
data = data.reset_index(drop=True)    

def remove_state_phase_one(value) :
    if value.lower() in d_state.keys():
        return 0
    else :
        if '&' in value :
            return 1 
        return 2
data['correction'] = data.SUBDIVISION.apply(remove_state_phase_one)
data_corrction = data[data.correction == 1]
data_corrction
```

```python
l = list(data_corrction.SUBDIVISION)
l_remove = []
for val in l:
    ind = val.split("&")
    d = data_corrction[data_corrction.SUBDIVISION == val]
    l_remove.append(d.index)
    for i in ind :
        i = i.strip()
        d.SUBDIVISION = i
        d.correction = 2
        data = data.append(d)
data = data.reset_index(drop=True)    
for i in l_remove:
    data = data.drop(i)
data = data.reset_index(drop=True)    
```

```python
def remove_state_phase_two(value) :
    if value.lower() in d_state.keys():
        return 0
    else :
        if ' ' in value :
            return 2
        return 3
data['correction'] = data.SUBDIVISION.apply(remove_state_phase_two)
data
data_corrction = data[data.correction == 2]
data_corrction
```

```python
l = list(data_corrction.SUBDIVISION)
l_remove = []
for val in l:
    ind = val.split(" ")
    d = data_corrction[data_corrction.SUBDIVISION == val]
    l_remove.append(d.index)
    for i in ind :
        i = i.strip()
        d.SUBDIVISION = i
        d.correction = 2
        data = data.append(d)
data = data.reset_index(drop=True)
for i in l_remove:
    data = data.drop(i)
data = data.reset_index(drop=True) 
data
```

```python
def remove_state_phase_three(value) :
    if value.lower() in d_state.keys():
        return 0
    else :
        return 3
data['correction'] = data.SUBDIVISION.apply(remove_state_phase_three)
data_corrction = data[data.correction == 0]
data_corrction
```

```python
final_states = data_corrction.groupby('SUBDIVISION').mean().copy()
final_states = final_states.iloc[:,:-1]
print(final_states)
data_corrction = data[data.correction != 0]
data_corrction
```

```python
l = final_states.index
l = [x.lower() for x in l]
left_state = []
for x in d_state.keys() :
    if x not in l :
        left_state.append(x)
dual_state = []
for val in left_state:
    if " " in val :
        i = val.split(" ")
        if i[0] == "West" :
            dual_state.append(i[1])
            continue
        dual_state.append(i[0])

def remove_state_phase_four(value) :
    if value.lower() in dual_state:
        return 0
    else :
        return 3
data['correction'] = data.SUBDIVISION.apply(remove_state_phase_four)
data_corrction = data[data.correction == 0]

def label_correction(value) :
    if value.lower() == "madhya":
        return "Madhya Pradesh"
    elif value.lower() == "bengal":
        return "West Bengal"
    elif value.lower() == "uttar" :
        return "Uttar Pradesh"
    elif value.lower() == "andhra" :
        return "Andhra Pradesh"
    else :
        return "0" 

data.SUBDIVISION = data.SUBDIVISION.apply(label_correction)
data_corrction = data_corrction[data_corrction.SUBDIVISION != "0"]
data_corrction
```
