```python
import pandas as pd

d_state = {'rajasthan': 0, 'madhya pradesh': 1, 'maharashtra': 2, 'jammu & kashmir': 3, 'uttar pradesh': 4, 'andhra pradesh': 5, 'gujarat': 6, 'karnataka': 7, 'orissa': 8, 'chhattisgarh': 9, 'tamil nadu': 10, 'bihar': 11, 'arunachal pradesh': 12, 'assam': 13, 'jharkhand': 14, 'west bengal': 15, 'himachal pradesh': 16, 'uttarakhand': 17, 'punjab': 18, 'haryana': 19, 'kerala': 20, 'manipur': 21, 'meghalaya': 22, 'mizoram': 23, 'nagaland': 24, 'tripura': 25, 'sikkim': 26, 'goa': 27, 'delhi': 28}

database = pd.read_csv('year_2017_individual_record.csv')
database = database.iloc[:,1:]
database
```

```python
def remove_state_phase_one(value) :
    if value.lower() in d_state.keys():
        return 0
    else :
        if '&' in value :
            return 1
        return 2
database['correction'] = database.SUBDIVISION.apply(remove_state_phase_one)
database_correction = database[database.correction == 1]
# database_correction
```

```python
def implementation_of_phase_one(value):
    val = value.split("&")
    return val[0]

def implementation_of_phase_one_2(value):
    val = value.split("&")
    return val[1]

database_2 = database_correction.copy()
database_correction.SUBDIVISION = database_correction.SUBDIVISION.apply(implementation_of_phase_one)
database_2.SUBDIVISION = database_2.SUBDIVISION.apply(implementation_of_phase_one_2)
database_correction = database_correction.append(database_2)
database = database[database.correction != 1]
database = database.append(database_correction).copy()
database.reset_index(drop=True,inplace=True)
# database
```

```python
l = ['West', 'East', 'North ', 'South', 'Gangetic', 'Region', 'Coastal', 'Sub Himalayan', 'Interior']
def remove_unwanted_words(value) :
    for c_word in l:
        if c_word in value:
            value = value.replace(c_word,"")
            print("found",c_word,value)
    return value.strip()

database.SUBDIVISION = database.SUBDIVISION.apply(remove_unwanted_words)

def remove_state_phase_two(value) :
    if value.lower() in d_state.keys():
        return 0
    else :
        if ' ' in value :
            return 1
        return 3
database['correction'] = database.SUBDIVISION.apply(remove_state_phase_two)
database_correction = database[database.correction == 1]
```

```python
label_s = database_correction.SUBDIVISION
split_able = {}
for s in label_s:
    in_s = s.split(" ")
    for s_v in in_s:
        if s_v.strip().lower() in d_state.keys():
            if s in split_able.keys():
                split_able[s] += [s_v] 
            else :
                split_able[s] = [s_v]
new_data = pd.DataFrame()
for i in split_able.keys() :
    d = database_correction[database_correction.SUBDIVISION == i]
    for v in split_able[i] :
        d.SUBDIVISION = v
        d.correction = 0
        new_data = new_data.append(d, ignore_index=True).copy()
new_data

        
```

```python
database = database[database.correction != 1]
database = database.append(new_data).copy()
```

```python
database.reset_index(drop=True, inplace=True)
database = database.groupby('SUBDIVISION', as_index=False).mean().copy()
database.reset_index(drop=True, inplace=True)
database
```

```python
database.loc[4,'SUBDIVISION'] = 'West Bengal'
database.loc[4,'correction'] = 0
print(database)
l = list(database.SUBDIVISION)
l = [x.lower() for x in l]
left_over = []
for item in d_state.keys():
    if item not in l:
        left_over.append(item)
print(left_over)
new_data = pd.DataFrame()      
for val in left_over:
    d = database[database.SUBDIVISION == 'Tripura'].copy()
    d.SUBDIVISION = val
    d.correction = 0
    new_data = new_data.append(d).copy()

database = database.append(new_data).copy()
database = database[database.correction == 0]
database.reset_index(drop=True, inplace=True)
```

```python
database = database.iloc[:,:-1]
database
database.to_csv('2017_filtered_records.csv')
```
