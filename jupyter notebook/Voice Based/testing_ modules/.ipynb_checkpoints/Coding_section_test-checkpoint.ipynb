```python
import tkinter as tk
import threading 
w_modes = ['file', 'coding', 'run', 'selection', 'scroll', 'delete']

working_mode = 'coding'

action = ""

code = ""
```

```python
def call(say,isString=False):
    global action
    canvas_action.itemconfig("command_view_txt",text=write_multiline_txt(say,30))
    if isString:
        while action == "":
            continue
        val = action
        action_txt.delete(0, tk.END)
        action = ""
        return val
    else :
        while action == "":
            continue
        val = int(action)
        action_txt.delete(0, tk.END)
        action = ""
        return val

def create_dataframe(n=None):
    """ Major var to be returned 
        n = number of columns
        column_name = list with column names
        var name
    """
    statement = "import pandas as pd \n\n"
    column_name = []
    var_name = 'dataframe'
    # if the number of columns are not defined get them
    if n == None :
        n = call("How many columns should the dataframe have ? \n")
    column_name = ['column_'+str(x) for x in range(n)]
    # asking for names to replace default columns names
    c = call("Would u like to change column names? y : n\n", True)
    if c == 'y' :
        column_name = []
        for i in range(n):
            column_name.append(call("Change column_" + str(i) + " to : ", True))
        print(column_name)
    # asking for change in var name if they want
    c = call("Would u like to change variable name for dataFrame ? y : n \n", True)
    if c == 'y' :
        var_name = call("new Variable name : ", True)
        
    statement += var_name + " = pd.DataFrame( columns=" + str(column_name) + ")"
    print(statement)
    return statement

def sub_dataframe_creator(canvas_action, canvas_coding, code,n=None):
    code = create_dataframe(n)
    refresh_codie_lines(canvas_coding, code)

```

```python
def command_analyser(command):
    global code,action
    print(command)
    try :
        c = command
        n = int(command[-9])
        command = command[:-9] + command[-8:]
        command = command.replace(" ","")
        print(command, n)
        if command == "createdataframewithcolumns":
            th = threading.Thread(target=sub_dataframe_creator, args=(canvas_action, canvas_coding, code,n))
            th.start()
            action_txt.delete(0, tk.END)
        else :
            action = c
    except:
        command = command.replace(" ","_")
        if command == "create_new_dataframe":
            th = threading.Thread(target=sub_dataframe_creator, args=(canvas_action, canvas_coding, code, None))
            th.start()
            action_txt.delete(0, tk.END)
        else :
            action = command
        
def write_multiline_txt(txt, limit):
    length = len(txt)
    s = ""
    for i,w in enumerate(txt):
        if i % limit == 0:
            s += '\n'
        s += w
    return s
```

```python
window = tk.Tk()

def refresh_codie_lines(canvas_coding, code):
    print("thread code : ", code)
    code_pat = code.split('\n')
    for i,loc in enumerate(code_pat):
        canvas_coding.itemconfig("line_of_code_"+str(i+1),text=loc)

def create_rounded_rect(canvas, x1, y1, x2, y2, **kwargs ) :
    h = y2 - y1
    canvas.create_oval( x1, y1, x1+h, y1+h, outline="", **kwargs )
    canvas.create_line( x1+h/2, y1 + h/2, x2-h/2, y1 + h/2, width=h, **kwargs)
    canvas.create_oval( x2 - h, y2 - h, x2, y2, outline="", **kwargs )

def call_command(event) :
    action = event.widget.get().lower().strip()
    command_analyser(action)
    
parent_canvas = tk.Canvas(window, width=1000, height=750)

canvas_action = tk.Canvas(parent_canvas, width=250, height=150)
parent_canvas.create_window(730, 580, anchor=tk.NW, window=canvas_action)
action_txt = tk.Entry(canvas_action, width=25)
action_txt.bind('<Return>', call_command)
canvas_action.create_text(125, 60, text="Data after processing view here", fill="#666666", tags="command_view_txt")
canvas_action.create_window((10,123), window=action_txt, anchor='nw')


# code for coding
canvas_coding = tk.Canvas(parent_canvas, width=700, height=730, bg="#eeeeee")
parent_canvas.create_window(5, 5, window=canvas_coding, anchor=tk.NW)

for i in range(20):
    canvas_coding.create_line(0, 40 + i*40, 710, 40 + i*40, width=20, fill="#cccccc")
    # index lines
    canvas_coding.create_text(15, 20 + i*40 , text=str(i*2 + 1), fill="#666666", font="Time 12")
    canvas_coding.create_text(15, 40 + i*40 , text=str(i*2 + 2), fill="#ffffff", font="Time 12")
    # coding lines
    canvas_coding.create_text(35, 13 + i*40 , text=" ", fill="#222222", font="Time 12", anchor=tk.NW, tags="line_of_code_" + str(i*2 + 1))
    canvas_coding.create_text(35, 33 + i*40 , text=" ", fill="#222222", font="Time 12", anchor=tk.NW, tags="line_of_code_" + str(i*2 + 2))
    
canvas_coding.create_line(28, 0, 28, 735, width="2", fill = "#ffffff")

canvas_status = tk.Canvas(parent_canvas, width=100, height=37)
create_rounded_rect(canvas_status, 5, 6, 90, 30, fill="#dddddd")
txt_status = canvas_status.create_text(50, 17, text=working_mode.capitalize(), font="Time 14", fill="#666666")
parent_canvas.create_window(890, 10, anchor=tk.NW, window=canvas_status)

parent_canvas.pack()
tk.mainloop()

print(code)
```
